```python
!pip install langchain-community
from langchain_community.document_loaders import WebBaseLoader
#Accept the Url and extract the relevant data usually known as web scrapping

loader = WebBaseLoader("https://www.geeksforgeeks.org/jobs/dna-growth-data-specialist-power-bi")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
        """
        ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:important keys role,experience,skills and description
        ### VALID JSON (NO PREAMBLE):
        """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':page_data})
type(res.content)
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
#convert textual data into json format like a dictionary object
json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```
